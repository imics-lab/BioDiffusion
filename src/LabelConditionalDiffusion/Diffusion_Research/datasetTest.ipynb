```python
mnist = MNIST_Conditional(oneD = False)
```

```python
dataloader = torch.utils.data.DataLoader(dataset=mnist, batch_size=1, shuffle=False)
```

```python
len(dataloader)
```

```python
mnist.org_data.size()
```

```python
dataloader
```

```python
i = 0
for dic in dataloader:
    org_img = dic['org_img'] 
    org_label = dic['org_label']
    target_img = dic['target_img']
    target_label = dic['target_label']
    cond_label = dic['cond_label']
    i += 1
    if i > 1:
        break
```

```python
print(org_label)
```

```python
from mpl_toolkits.axes_grid1 import ImageGrid


fig = plt.figure(figsize=(5, 10))
grid = ImageGrid(fig, 111, nrows_ncols=(1, 2), axes_pad=0.1)

for ax, im in zip(grid, [org_img, target_img]):
    ax.imshow(im[0], cmap='gray')
    ax.axis('off')
print(org_label, target_label, cond_label)

plt.show()
```

```python
from UNIMIB import *
```

```python
unimib = unimib_load_dataset(is_normalize = True)
```

```python
from torch.utils import data
import matplotlib.pyplot as plt
```

```python
dataloader = data.DataLoader(unimib, batch_size=1, num_workers=1, shuffle = True)
```

```python
fig, axs = plt.subplots(1, 10, figsize=(30,3))
fig.suptitle('UniMiB Data samples', fontsize=25)
for i, (signal, label) in enumerate(dataloader):
    if i < 10:
        axs[i].plot(signal[0][0][:])
        axs[i].plot(signal[0][1][:])
        axs[i].plot(signal[0][2][:])
    else:
        break
```

```python
unimib_running = unimib_load_dataset(is_normalize = True,  single_class = True, class_name= 'Running',)
```

```python
Runningloader = data.DataLoader(unimib_running, batch_size=1, num_workers=1, shuffle = True)
```

```python
fig, axs = plt.subplots(1, 10, figsize=(30,3))
fig.suptitle('UniMiB Running', fontsize=25)
for i, (signal, label) in enumerate(Runningloader):
    if i < 10:
        axs[i].plot(signal[0][0][:])
        axs[i].plot(signal[0][1][:])
        axs[i].plot(signal[0][2][:])
    else:
        break
```
